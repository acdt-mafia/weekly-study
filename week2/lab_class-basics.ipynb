```python
"""
General guide: 이 .ipynb 파일 전체에서 이 코드 블록 하나만을 수정하여 이하 블록들의 요구사항을 만족시켜야 합니다.
"""

class Point():
    # 아래 pass를 지우고 클래스를 설계하세요.
    def __init__(self, x, y):
        self.x = x
        self.y = y
        # todo 1을 위한 코드= 
        print(f"Point initiated: ({self.x}, {self.y})")
        ## 1이랑 2 둘 다 만족시킬 수 있는 방법을 못찾았어요. 아래는 todo2를 위한 코드입니다
        #print(f" point info: x={self.x}, y={self.y} ")

    def __str__(self):
        return f"Point info: x={self.x}, y={self.y}"
    
    def __repr__(self):
        return self.__str__()
    
    def __eq__(self, other):
        return True if self.x == other.x and self.y == other.y else False
```

# To Do 1 노트
__init__ 은 파이썬에서 클래스의 생성자 메소드. 
생성자는 뭐고 메소드는 뭐냐. 

메소드는 클래스 안에 들어있는 함수. 
생성자는 인스턴스가 만들어지는 순간 자동으로 호출되는 메소드. 

set 이랑 init의 차이
set은 필요할 때 수동으로 키는 함수고, 상태 변경이 목적이다. 
반면 init은 생성될 때 자동으로 켜지며 기본 상태 보장을 목적으로 함. 
init을 안 쓰면 객체가 불완전해짐.

그렇다면 __repr__은 뭘까?
