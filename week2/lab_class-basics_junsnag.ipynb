```python
"""
General guide: 이 .ipynb 파일 전체에서 이 코드 블록 하나만을 수정하여 이하 블록들의 요구사항을 만족시켜야 합니다.
"""

class Point():
    def __init__(self, x, y):
        self.x = x
        self.y = y
        print(f"Point initiated: ({self.x},{self.y})")

    def __str__(self):
        return f"Point info: x={self.x}, y={self.y}"

    def __repr__(self):
        return f"Point info: x={self.x}, y={self.y}"

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y



    



        
```

```python
"""
To-Do 1: 아래 코드를 실행했을 때, 다음과 같이 순서대로 출력되어야 합니다.

Point initiated: (1, 2)
Point initiated: (1, -1)
Point initiated: (1, 2)
"""
point_1 = Point(x=1, y=2)
point_2 = Point(x=1, y=-1)
point_3 = Point(x=1, y=2)
```

```python
"""
To-Do 2: 아래 코드를 실행했을 때, 아래와 같이 순서대로 출력되어야 합니다.

Point info: x=1, y=2
Point info: x=1, y=-1
"""

print(point_1)
point_2
```

```python
"""
To-Do 3: 아래 코드를 실행했을 때, 아래와 같이 출력되어야 합니다.

True
"""
is_equal_1 = point_1==point_3
print(is_equal_1)
```
